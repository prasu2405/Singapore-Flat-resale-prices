Regression model which predict the resale_price


```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
import numpy as np
```

```python
def best_ML_algorithm(df,algorithms):
        x=df.drop(columns=['resale_price'],axis=1)
        y=df['resale_price']

        x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

        results={}

        for algorithm in algorithms:

                model=algorithm().fit(x_train,y_train)
                y_pred=model.predict(x_test)
                
                MAE=metrics.mean_absolute_error(y_test, y_pred)
                MSE=metrics.mean_squared_error(y_test, y_pred)
                RMSE=np.sqrt(metrics.mean_squared_error(y_test, y_pred))
                r2 = metrics.r2_score(y_test, y_pred)

                results[algorithm.__name__] = {'Mean Absolute Error':MAE,'Mean Squared Error':MSE,'Root Mean Squared Error':RMSE,
                                        'R² score':r2}
                
        best_algorithm = max(results.items(), key=lambda item: item[1]['R² score'])[0]  
        
        return best_algorithm, results
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
algorithms = [LinearRegression,DecisionTreeRegressor,RandomForestRegressor,AdaBoostRegressor,KNeighborsRegressor]
best_algorithm, results = best_ML_algorithm(new_df, algorithms)

results_df = pd.DataFrame(results).transpose()

print("**Evaluation Metrics for All Algorithms:**")
print(results_df.to_string())

print("**Best Algorithm:**",best_algorithm)
```

Random Forest Regressor has the highest R2 Score best algorithm 



```python
df_alg=new_df.copy()
```

```python
x=df_alg.drop('resale_price',axis=1)
y=df_alg['resale_price']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42)
model1=RandomForestRegressor(n_estimators=100,max_depth=30,min_samples_split=2,min_samples_leaf=1).fit(x_train,y_train)
y_predict=model1.predict(x_test)
# Evaluation Metrics:
print("MAE: {:.4f}".format(metrics.mean_absolute_error(y_test,y_predict)))
print("MSE: {:.4f}".format(metrics.mean_squared_error(y_test,y_predict)))
print("RMSE: {:.4f}".format(np.sqrt(metrics.mean_squared_error(y_test,y_predict))))
print("R2: {:.4f}".format(metrics.r2_score(y_test,y_predict)))
```
